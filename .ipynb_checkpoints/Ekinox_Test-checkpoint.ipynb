# Import libs

```python
# Data Management
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
pd.set_option('display.max_columns', 100)

# Time performance
from time import time

# home-made functions
import src.ad_hoc_functions as ahf

# Scikit learn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score, recall_score

import random
```

# data import and management

## Import and check formats

```python
# Import data 

data_math = pd.read_csv('student/student-mat.csv',sep=';')
data_math = data_math.rename(columns={"G1": "G1_math", "G2": "G2_math", "G3": "G3_math"})


data_port = pd.read_csv('student/student-por.csv',sep=';')
data_port = data_port.rename(columns={"G1": "G1_port", "G2": "G2_port", "G3": "G3_port"})

print("shape of data: {}".format(data_math.shape))
print(data_math.dtypes)
data_math.head()
```

```python
# apply 'add_info' function on Math and Portugese students : 
# Compute mean of the 3 grades, add a flag to students, and flag the ones who passed the test (target)

data_math = ahf.add_info(data_math, 'math')
data_port = ahf.add_info(data_port, 'port')

print("shape of data_math: {}".format(data_math.shape))
print("shape of data_port: {}".format(data_port.shape))

data_port.head()
```

We add through the merging function the following information :
- the studend took math classes or not

It will constitute an intersting add-in to the dataset

```python
# Get the info if student has a math grade
data = pd.merge(data_port,
                data_math[['school','sex','age','address','famsize','Pstatus','Medu','Fedu','Mjob','Fjob','reason','nursery','internet','flag_math','pass_math']],
                left_on = ['school','sex','age','address','famsize','Pstatus','Medu','Fedu','Mjob','Fjob','reason','nursery','internet'],
                right_on= ['school','sex','age','address','famsize','Pstatus','Medu','Fedu','Mjob','Fjob','reason','nursery','internet'],
                how='left',
                suffixes=('', '_y')
                     )

# Remove duplicated columns
data.drop(data.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)

# replace null values flag_math as 0 as 
data = data.fillna(0)

# print info
print("shape of data: {}".format(data.shape))
print(10*"*")
print(data.groupby('flag_math').size())

data.head()
```

## Check NAs and duplicates

### Count NAs

```python
# Check missing values per feature
print(data.isna().sum())
print(8*"-")

# check if some rows have multiple NAs
data['is_null_sum'] = data.isnull().sum(axis=1)
print(data.groupby('is_null_sum').size())
```

## Remove duplicates
We check if many rows are similar to each others

```python
# Check duplicates

duplicate_rows_data = data[data.duplicated()]
print("number of duplicate rows: ", len(duplicate_rows_data))

# Remove duplicates
data = data.drop_duplicates()

print(data.shape)
```

## Replace binary variables by 1 (yes) or 0 (no)

```python
data = data.replace({'yes':1, # binary variables
                     'no':0, 
                     'LE3':0, # Family size
                     'GT3':1, 
                     'F':1, # sex
                     'M':0, 
                     'U':1, # adress 
                     'R':0, 
                     'GP':0, # school
                     'MS':1, 
                     'T':1, #Pstatus
                     'A':0,}
                   )

# display the gead to check transformation
data.head()
```

## Get dummies for some categorical variables
* We suggest to get dummy variables for some of the categorical values: Mjob, Fjob, reason & guardian.


```python
# Drop_First allows to delete automatically the N-1 dummy variable and thus avoid colinearity

dummy_var = {'Mjob', 'Fjob', 'reason', 'guardian'}

dummies_data = pd.get_dummies(data[dummy_var],drop_first=True)

print("shape of data: {}".format(dummies_data.shape))
dummies_data.head()
```

## Cleaned Dataset

**To sum-up, here are actions we ran on the initial dataset:**

* We added to the G1, G2, G3 variables a suffix to mention to which class it is related
* We added one column : Math flag (binary : student took math classe or not)
* We compute mean of the two first grades (G1, G2)
* We flagged 1 or 0 the student based on the dataset in which we can find them
* We flagged student if they passed the exam or not (G3 >= 10) --> this will constitue our target
* We checked NAs
* We removed duplicates
* We reformat binary variables so that they turn to numeric
* We dummified categorical variables if they have more than 2 modalities

```python
# Merge cleaned data with dummies

data_training = pd.merge(data,
                         dummies_data,
                         left_index = True,
                         right_index = True,
                         how = 'inner')
data_training = data_training.drop({'Mjob', 'Fjob', 'reason', 'guardian', 'G1_port', 'G2_port', 'G3_port', 'Gmean_port'}, axis=1)


print("shape of data: {}".format(data_training.shape))
data_training.head()
```

## Scaling values

```python
# Define the scaler
scaler = MinMaxScaler()

# Fit the scaler on the data
scaler.fit(data_training)

# Transform the data and store results in a new dataframe
data_training_scaled = scaler.transform(data_training)
data_training_scaled = pd.DataFrame(data_training_scaled,columns=data_training.columns, index= data_training.index)

print("shape of data: {}".format(data_training_scaled.shape))
data_training_scaled.head()
```

# Modelization

## Train-Test Split
We set 2 different splits : one from scaled values, the other one for normal values. Tree-based models don't need scaled values as input

```python
# Define independant features (X) and target (y)

# From reduced dataset
y_reduced = data_training_scaled['pass_port']
X_reduced = data_training_scaled.drop(['pass_port'],axis=1)

# From initial dataset
y = data_training['pass_port']
X = data_training.drop(['pass_port'],axis=1)

print("shape of X: {}".format(y_reduced.shape))
print("shape of y: {}".format(X_reduced.shape))

```

```python
# Create the train and test sets (10%) ...

# ...for reduced dataset
X_train_R, X_test_R, y_train_R, y_test_R = train_test_split(X_reduced, y_reduced, test_size=0.10, random_state=1813)

# ... from initial dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1813)


for df, name in zip([X_train, X_test, y_train, y_test],['X_train', 'X_test', 'y_train', 'y_test']):
    print("shape of {}: {}".format(name,df.shape))
    
```

## Model 0 : RANDOM (baseline model)

```python
# create empty lists to fill them with random values
y_hat_proba = []
y_hat = []

# random values as proba and target to simulate a random
for i in range(1,len(X_test)+1):
    random_proba = random.uniform(0, 1)
    random_target = round(random_proba,0)
    
    y_hat_proba.append(random_proba)
    y_hat.append(random_target)
```

```python
# compute metrics

ll = log_loss(y_test, y_hat_proba)
accuracy = accuracy_score(y_test, y_hat)
recall = recall_score(y_test, y_hat)

print("LogLoss: {}".format(round(ll,2)))
print("Accuracy: {}%".format(round(accuracy * 100.0),2))
print("Recall: {}%".format(round(recall * 100.0),2))
```

```python
# plot matrices

ahf.plot_confusion_matrix(y_test, y_hat)
```

## Model I : logistic regression

I suggest to choose a logistic regression. It will give us a good idea of the results we can expect considering the data we have.

### Predict

```python
# Define the model and parameters
LR = LogisticRegression() # random_state=1703#807

# Fit the model
LR.fit(X_train_R,y_train_R)
```

```python
# Predict 
y_hat = LR.predict(X_test_R)
y_hat_proba = LR.predict_proba(X_test_R)

ll = log_loss(y_test_R, y_hat_proba)
accuracy = accuracy_score(y_test_R, y_hat)
recall = recall_score(y_test_R, y_hat)

print("LogLoss: {}".format(round(ll,2)))
print("Accuracy: {}%".format(round(accuracy * 100.0),2))
print("Recall: {}%".format(round(recall * 100.0),2))
```

```python
# Explore feature importances
feature_importances = pd.DataFrame(LR.coef_, columns=X_train_R.columns)
feature_importances
```

**Takeaways :**
* **School** : Gabriel Pereira seems to influence the results upwards
* **Failures** and **absences** influence grately downwards the 'chance' to pass the Portugese exam
* Globally, for **mother's and father's job** results are consistent with primary observations : no real impact
* Partying on week-end does not help...
* Taking **Math classes** does not help to pass Portugese exam. However **passing math exam** is a good sign
* **Wishing to pursue scholarship** beyond these classes is a good sign as well

```python
ahf.plot_confusion_matrix(y_test_R, y_hat)
```

## Model II : Random Forest
* First check different depth configurations to identify overfitting
* Check different parameters through a Gridsearch to identify the best model

### Check Overfitting

```python
# There are about 30 attributes in the dataset. Let's choose 20 as the max depth we will test
depth_values = [i for i in range(1,20)]
```

```python
# Write a quick loop that will iterate on each depth

train_scores = list()
test_scores = list()

for i in depth_values:
    
    # configure the model
    RF_model = RandomForestClassifier(max_depth=i)
    
    # fit model
    RF_model.fit(X_train, y_train)
    
    # evaluate on the train dataset
    yhat_train = RF_model.predict(X_train)
    train_acc = accuracy_score(y_train, yhat_train)
    train_scores.append(train_acc)
    
    # evaluate on the test dataset
    yhat_test = RF_model.predict(X_test)
    test_acc = accuracy_score(y_test, yhat_test)
    test_scores.append(test_acc)
    
sns.lineplot(x= depth_values, y= train_scores)
sns.lineplot(x= depth_values, y= test_scores)    
```

We estimate that RF will **overfit** if we go above a depth of 5. Let's run a GS with other paramters to find best model

### Run a gridsearch to find best parameters

```python
# Difine research perimter
parameters = {'n_estimators':[50,100,150],'max_depth':[1,2,3,4,5], 'random_state':[1736]}

# Define the gridsearch with above paramters
GS_RF = GridSearchCV(RandomForestClassifier(), param_grid=parameters, cv=5)
```

```python
# Run the predictions

time_0 = time()

GS_RF.fit(X_train,y_train)

print("it took: {} seconds".format(round((time()-time_0),2)))
```

```python
# best_esti = paramtres du modle qui a le mieux fonctionn
best_esti_clf = GS_RF.best_estimator_
best_esti_clf
```

```python
# best_sc = rsultats du meilleur modle
best_sc_clf = GS_RF.best_score_
best_sc_clf
```

```python
# Predict 
y_hat = best_esti_clf.predict(X_test)
y_hat_proba = best_esti_clf.predict_proba(X_test)

ll = log_loss(y_test, y_hat_proba)
accuracy = accuracy_score(y_test, y_hat)
recall = recall_score(y_test, y_hat)

print("LogLoss: {}".format(round(ll,2)))
print("Accuracy: {}%".format(round(accuracy * 100.0),2))
print("Recall: {}%".format(round(recall * 100.0),2))
```

```python
# Feature Importance
feature_importances = pd.DataFrame(best_esti_clf.feature_importances_, index = X_train.columns, columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```

```python
ahf.plot_confusion_matrix(y_test, y_hat)
```

**Takeaways :**
- Feature importance is consistent with LR outputs
